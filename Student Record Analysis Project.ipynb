```python
#importing the Libraries needed 


import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#reading the imported csv file

df = pd.read_csv("student_scores.csv")

print(df.head(10))
```

```python
df.describe()
```

```python
#checking the column datatypes
df.info()
```

```python
#checking total sum of null values for each column

df.isnull().sum()
```

```python
# drop the unnamed column

df = df.drop("Unnamed: 0", axis = 1)
print(df.head(10))
```

```python
# Change weekly study hours column 

df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05 - Oct", "5 - 10")
df.head()
```

```python
#gender Distribution
plt.figure(figsize = (6, 6))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

## Analysis from the above chart states that the number of females in the data is more than the number of males.

```python
gb = df.groupby("ParentEduc").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore" : "mean"})
print(gb)
```

```python
plt.figure(figsize = (6, 6))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

## The heatmap shows that parents who are educated have a positive impact on their student scores.

```python

# Lets see the impact of parents marital status on Student Scores

gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore" : "mean"})
print(gb1)
```

```python
plt.figure(figsize = (6, 6))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

## The heatmap proves that the marital status of parents do not have a negative impact on the student scores.

##  Distribution of Ethnic Groups

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

# print(groupA["EthnicGroup"])
l =  ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
